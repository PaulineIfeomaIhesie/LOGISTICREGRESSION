```python
#if the target shows 0,1 or true or false is logistic regression because its classified.
```

```python
#convert the sex column(categorical or string column) from string to numerical column

#a map is a function that convert a categorical or  string column to a numerical column

#you can use other functions like cat(cat converts each values to numerical)
Titanic['Sex'].map({'male': 0, 'female': 1})
```

```python
Titanic['Sex']=Titanic['Sex'].map({'male': 0, 'female': 1})
```

```python
Titanic
```

```python
#extract the columns you are going to use for your prediction
#x is independent , y is dependent

x = Titanic[["Pclass", "Sex", "Age", "SibSp", "Parch", "Fare"]]
y = Titanic["Survived"]
```

```python
# we need to split the dataset to enbale us predict the dataset

#split the x and y

```


# split the x and y

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(x,y, test_size=0.2, random_state=0)
```

```python
#fit the model(train)
```

```python
from sklearn.linear_model import LogisticRegression
```

# create an instance for LogisticRegression



```python
model = LogisticRegression()
```

```python
model.fit(x_train, y_train)
```

```python
a = model.predict(x_train)
```
